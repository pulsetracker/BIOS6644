```python
##################################################################################################################
#
# YouDo:
#  1) Remove any rows with na values
# 
#  2) Split the data into "test" and "train" bits with ~30% in the test set.
#       The "Label" column is the "y" data.  All other columns are the "X" data
#       Give them these names:
#       X_train, X_test, y_train, y_test

#######################################  BEGIN STUDENT CODE  #####################################################

# 1)
#df2.info()
#I think we got rid of all the Na earlier?
#AFter looking at the data below there is a NaN in row 33!
df3 = df2.dropna() #this should drop any rows containing Na 
df3
#df3.info()

 
#2) The following is how I set it up.  
y_train = df3.Label[0:33]
y_test = df3.Label[34:]
X_train = df3[0:33]
X_test = df3[34:]

#print(y_train, y_test, X_train, X_test)

# I found this online. https://www.datacamp.com/community/tutorials/random-forests-classifier-python
# Not sure how to get it set up completely this way. 
#X=data[df3[1:33]  # Features
#y=data[df3('Label')]  # Labels
# Using this example - I am not sure how to designate the two sets of data properly. 
# Split dataset into training set and test set
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test
#######################################   END STUDENT CODE   #####################################################
##
##
##
## Instructor comment
## That's the right way to do it.   
## "X" is the data and "y" is the label


from sklearn.model_selection import train_test_split
y = df3.Label
X = df3[df3.columns.drop('Label')]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

```

```python
#Build a model to classify the Label based on the other columns

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=8675309)
clf.fit(X_train, y_train)
#X_train
```

```python
# Test the model
predictions = clf.predict(X_test)
reality = y_test
#print(predictions)
#print(clf.feature_importances_)
#print(clf.predict([[0,0,0,0]]))

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
#print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Accuracy:",metrics.accuracy_score(reality, predictions))
print(metrics.classification_report(reality, predictions))

```
