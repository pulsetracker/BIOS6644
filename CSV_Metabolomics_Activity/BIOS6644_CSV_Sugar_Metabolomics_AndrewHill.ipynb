```python
##################################################################################################################
#
# YouDo:
# 
#    1) Get rid of any columns that have only NaN values
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html
#
#######################################  BEGIN STUDENT CODE  #####################################################

df = df.loc[:,~np.all(np.isnan(df),axis=0)]

#######################################   END STUDENT CODE   #####################################################

assert df.shape == (92, 125)
```

```python
##################################################################################################################
#
# YouDo:
# Make a pair of plots as follows:
#    1) Show the correlation matrix between columns of the data
#        (hint 1: it should be 124x124, hint 2: check out plt.imshow() 
#         Hint 2: numpy has a function for generating the correlation matrix)
#
#    2) Make a similar plot showing correlation > 90%
#
#  Stretch goal:  using plt.subplot()  make a single figure with both plots side-by-side
#
#######################################  BEGIN STUDENT CODE  #####################################################
# Make both figures
f,axes = plt.subplots(nrows=2,figsize=(8,16))
# Correlation from Pandas DataFrame
corr_df = df.corr()
# Show the image and set title
im_1 = axes[0].imshow(corr_df)
axes[0].set_title("Correlation Matrix of all Components")
# Get location of first image
right_of_first = make_axes_locatable(axes[0])
# Add a colorbar to show values.
colorbar_1_axis = right_of_first.append_axes("right",size="5%",pad=0.1)
f.colorbar(im_1,cax=colorbar_1_axis)


# Get all instances of correlation over 90%
g90_idcs = (corr_df > 0.9).values
# Diagonal (self-correlation) doesn't count
d_idcs = np.diag_indices_from(g90_idcs)
g90_idcs[d_idcs] = False
g90_idcs = np.any(g90_idcs,axis=0)
# Get filtered matrix
reduced_mat = corr_df.values[g90_idcs]
reduced_mat = reduced_mat[:,g90_idcs]
# Show image
im_2 = axes[1].imshow(reduced_mat)
# Get location of second image
right_of_second = make_axes_locatable(axes[1])
colorbar_2_axis = right_of_second.append_axes("right",size="5%",pad=0.1)
f.colorbar(im_2,cax=colorbar_2_axis)
# Set title
axes[1].set_title("Correlation Matrix of Components with Corr. > 90%")
# Set labels for the second image.
lbls = list(corr_df[g90_idcs].index)
axes[1].set_yticks(range(reduced_mat.shape[0]))
axes[1].set_xticks(range(reduced_mat.shape[0]))
axes[1].set_xticklabels(lbls,ha="right")
axes[1].set_yticklabels(lbls)
axes[1].tick_params("x",labelrotation=50)

#######################################   END STUDENT CODE   #####################################################
```

```python
##################################################################################################################
#
# YouDo:
#  1) Remove any rows with na values
# 
#  2) Split the data into "test" and "train" bits with ~30% in the test set.
#       The "Label" column is the "y" data.  All other columns are the "X" data
#       Give them these names:
#       X_train, X_test, y_train, y_test

#######################################  BEGIN STUDENT CODE  #####################################################

# 1)
non_na_rows = ~np.any(np.isnan(df),axis=1)
filt_df = df[non_na_rows]


#2)
test_vals = np.random.choice(filt_df.shape[0],int(0.3*filt_df.shape[0]),replace=False)
train_vals = np.setdiff1d(list(range(filt_df.shape[0])),test_vals)
assert(len(test_vals) + len(train_vals) == filt_df.shape[0])

X_train = filt_df.iloc[train_vals,1:]
y_train = filt_df.iloc[train_vals,0]
X_test = filt_df.iloc[test_vals,1:]
y_test = filt_df.iloc[test_vals,0]
#######################################   END STUDENT CODE   #####################################################
```

```python
#Build a model to classify the Label based on the other columns

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=8675309)
clf.fit(X_train, y_train)
```

```python
# Test the model
predictions = clf.predict(X_test)
reality = y_test
```

```python
# YouDo:
#  1) Evaluate the "goodness" of the predictions and write a paragraph describing the results as a markdown.
#
#######################################  BEGIN STUDENT CODE  #####################################################

from sklearn.metrics import classification_report

print("Train Data Classification Result")
print(classification_report(y_train,clf.predict(X_train)))

print("Test Data Classification Result")
print(classification_report(y_test,predictions))

#######################################   END STUDENT CODE   #####################################################
```

# Analysis
Based on the results above, it seems that this classifier is better at classifying Class 1 than Class 2. On test data, the model is able to achieve an F1 Score of 0.75 on Class 1, while it only achieves a F1 Score of 0.43 on Class 0. This could be because there are more examples of Class 1 in our Test data than Class 0. We can verify this by running a histogram on the Label column. Because there are more examples of Class 1, the model is probably more likely to choose it during classification, meaning better performance on Class 1 at the cost of worse performance on Class 0.
