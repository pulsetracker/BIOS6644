```python
#######################################  BEGIN STUDENT CODE  #####################################################
import numpy as np
#2) Visualization code
# If in doubt, use matplotlib.   If you want to get fancy, use bokeh, plotly, etc.
fig,ax = plt.subplots(figsize=(6,5))
ax.set_xlabel("Total Population")
ax.set_ylabel("Percentage Over 24 with HSD or Higher")
ax.set_title("% of Pop. with High School Diploma vs. Total Population")
#3) Statisical testing code (as appropriate)
lm = stats.linregress(df.Population_Total,df.Percent_Education_Population_AgeOver24_HSD_Higher)
print("P-value of Linear Regression coefficient: %0.4f" % lm[3])
ax.plot(df.Population_Total,df.Population_Total*lm[0] + lm[1],"r")
ax.scatter(df.Population_Total,df.Percent_Education_Population_AgeOver24_HSD_Higher)
ax.scatter(df.Population_Total.mean(),df.Percent_Education_Population_AgeOver24_HSD_Higher.mean())
ax.legend(["Linear Model Fit","Data Points","Mean Point"])
ax.text(0.7,0.2,"P-value: %0.4f\n$R^2$ value: %0.4f" % (lm[3],lm[2]),transform=ax.transAxes)
#######################################   END STUDENT CODE   #####################################################
```

## Analysis

It seems like my hypothesis is wrong. There does seem to be some positive trend between Population and %HSD in the data, but the vast majority of tracts have a population around ~4000, and very high rates of High School Graduation, which means that any trend would be overpowered by the majority of the data points that don't follow it and not show up in the regression line. 

The actual regression line seems to show very little correlation between these two data points, and the P-Value is predictably bad as well.
