## Analysis
A pearson correlation was performed for those over the age of 24 with no High school diploma compared to the population density of the county.  The pearson r = 0.207 with a p value = 1.54e-13.  This shows a small positive correlation between population density in the county and having no high school diploma.  Comparing the correlation between those people with a high school diploma or higher results in a pearson r = -.207 with a p value = 1.49e-13.   This pearson r value indicates a small negative correlation with population density and having a high school diploma or higher.  Neither effect is very great.
