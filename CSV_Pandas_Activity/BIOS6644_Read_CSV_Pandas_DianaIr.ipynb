```python
#2) Visualization code
# If in doubt, use matplotlib.   If you want to get fancy, use bokeh, plotly, etc.

# library & dataset
datafile = '/home/dir0417/Desktop/BIOS6644-master/Data/Educational_Attainment_Census_Tracts.csv'
df = pd.read_csv(datafile)

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cmx
from pandas import read_csv
```

```python
#Scatter of % of edu, >24 y/o with BS degree
x=df['OBJECTID']
y=df['Percent_Education_Population_AgeOver24_Bachelors_Higher']

# Get unique names of counties
uniq = list(set(df['County']))


# Set the color map to match the number of counties
z = range(1,len(uniq))
hot = plt.get_cmap('nipy_spectral')
cNorm  = colors.Normalize(vmin=0, vmax=len(uniq))
scalarMap = cmx.ScalarMappable(norm=cNorm, cmap=hot)

# Plot each counties
for i in range(len(uniq)):
    indx = df['County'] == uniq[i]
    plt.scatter(x[indx], y[indx], s=10, color=scalarMap.to_rgba(i), label=uniq[i])

plt.xlabel('population')
plt.ylabel('%ed_>24')
plt.title('% of people, 24 years of age or older, with Bachelors or higher in each county ')
plt.legend(loc='upper left')
plt.show()
```

```python
# Diana's interpretation:
#     I kind of need to play with the color a bit, but from this chart, since I am interested in comparing education
#     level of Douglas County to other counties.  I can safely say that Douglas is one of the middle blue dots.  
#     I will change the colors now that we know Douglas is one of the middle blue dots.

## Instructor comment.
## One trick I use in this kind of situation is actually to plot the point of interest twice.   In this case, 
## plot all of the dots as normal and then use a second plt.scatter() line to just plot the Douglas county dot 
## with a unique symbol.
## Also, while I like to have a legend, this one is a bit over the top--or should we say, "out the bottom"
```

```python
#Scatter of % of edu, >24 y/o with BS degree
x=df['OBJECTID']
y=df['Percent_Education_Population_AgeOver24_Bachelors_Higher']

# Get unique names of counties
uniq = list(set(df['County']))

# Set the color map to match the number of counties
z = range(1,len(uniq))
hot = plt.get_cmap('flag')
cNorm  = colors.Normalize(vmin=0, vmax=len(uniq))
scalarMap = cmx.ScalarMappable(norm=cNorm, cmap=hot)

# Plot each counties
for i in range(len(uniq)):
    indx = df['County'] == uniq[i]
    plt.scatter(x[indx], y[indx], s=10, color=scalarMap.to_rgba(i), label=uniq[i])

plt.xlabel('population')
plt.ylabel('%ed_>24')
plt.title('% of people, 24 years of age or older, with Bachelors or higher in each county ')
plt.legend(loc='upper left')
plt.show()
```
