```python
#######################################  BEGIN STUDENT CODE  #####################################################

#2) Visualization code
# If in doubt, use matplotlib.   If you want to get fancy, use bokeh, plotly, etc.

''' The first thing I would like to point out is the y axis label does not correctly describe the actual data being shown. 
The y axis label says we are looking at the percent of the population that has higher than a BS degree when 
in reality we are looking at the percent of the population that does not have a HS diploma. 

I redid the plot to display the percent of the population with Higher than a BS degree'''

# I want to add a trendline to the scatterplot to determine what the trend between 
# the percent of the population with a BS and the population density
plt.figure(figsize = (6,4))
plt.plot(df.Population_Density_PerLandSquareMile, df.Percent_Education_Population_AgeOver24_Bachelors_Higher,'.')
plt.xlabel('Population Density')
plt.ylabel('% of Population >= BS Degree')
plt.title('Percent of Population with >= BS Degree vs the Population Density')
z = np.polyfit(df.Population_Density_PerLandSquareMile, df.Percent_Education_Population_AgeOver24_Bachelors_Higher, 1)
p = np.poly1d(z)
plt.plot(df.Population_Density_PerLandSquareMile,p(df.Population_Density_PerLandSquareMile),"r--")
```

```python
#3) Statisical testing code (as appropriate)
# scipy.stats has most of what you might need
from scipy import stats

# Pearson Correlation between Pop Denisty and % Education with more than BS
stats.pearsonr(df.Population_Density_PerLandSquareMile, df.Percent_Education_Population_AgeOver24_Bachelors_Higher)
```

```python
stats.pearsonr(df.Population_Total, df.Percent_Education_Population_AgeOver24_Bachelors_Higher)
```

```python
stats.pearsonr(df.Population_Total, df.Percent_Education_Population_AgeOver24_HSD_Higher)
```

```python
# I want to see if there is a statistical significance between high populations and low populations. 
# According to the summary statistics table shown above, the largest population is 15,869. 
# Therefore, I will define low population as being below 7,935 (half of 15,869). High Population is greater than 7,935.
low_pop_df = df.copy()
high_pop_df=df.copy()

low_pop_df = low_pop_df[low_pop_df.Population_Total < 7935]
low_pop_df.shape
```

```python
low_pop_df.describe()
```

```python
high_pop_df = high_pop_df[high_pop_df.Population_Total >= 7935]
high_pop_df.shape
```

```python
high_pop_df.describe()
```

```python
# Is there a difference between the percent of population with BS or higher in high and low populations?
stats.ttest_ind(high_pop_df.Percent_Education_Population_AgeOver24_Bachelors_Higher,low_pop_df.Percent_Education_Population_AgeOver24_Bachelors_Higher)
```

```python
# Is there a difference between the percent of population with HSD or higher in high and low populations?
stats.ttest_ind(high_pop_df.Percent_Education_Population_AgeOver24_HSD_Higher,low_pop_df.Percent_Education_Population_AgeOver24_HSD_Higher)
```

My hypothesis was definitely incorrect. This hypothesis was made on a plot that had been mislabled, therefore it would have been a total fluke if my hypothesis was correct. When I did further analysis, it turns out there is no relationship between population size and level of education. There is no correlation between population density and percent of population with more than a BS (r = 0.18), total population and percent of population with more than a BS (r = 0.05), or total population and percent of population with more than a high school diploma (r = 0).

To further prove to myself that there was no relationship between population and education, I performed a t test on the education values between high and low populations. The p value on the percent of the population with more than a BS was 0.40. The p value on the percent of the population with more than a HS diploma was 0.19. Both calculations support the fact that there is no statistically significant difference between high and low populations as far as education is concerned.
