## Analysis

Census tracts in Boulder county have on average 24.2% more people with a bachelor's degree or higher than other counties in the state. This difference is statistically significant with a p-value of <.0001. The stereotype is true: Boulder is actually quite well educated.

```python
#3) Statisical testing code (as appropriate)

# sourcing code below from: https://scientificallysound.org/2017/05/16/independent-t-test-python/    

from scipy import stats
from math import sqrt
from scipy.stats import t

# Run independent t-test
ind_t_test = stats.ttest_ind(df.Percent_Education_Population_AgeOver24_Bachelors_Higher[df.binary_boulder == 1],
                             df.Percent_Education_Population_AgeOver24_Bachelors_Higher[df.binary_boulder == 0])

# Calculate the mean difference and 95% confidence interval
N1 = 68 #which we know from an earlier code chunk
N2 = 1174 # which is 1242-68
degrees = (N1 + N2 - 2)
std1 = df.Percent_Education_Population_AgeOver24_Bachelors_Higher[df.binary_boulder == 1].std()
std0 = df.Percent_Education_Population_AgeOver24_Bachelors_Higher[df.binary_boulder == 0].std()
std_N1N2 = sqrt( ((N1 - 1)*(std1)**2 + (N2 - 1)*(std0)**2) / degrees) 

diff_mean = df.Percent_Education_Population_AgeOver24_Bachelors_Higher[df.binary_boulder == 1].mean() - df.Percent_Education_Population_AgeOver24_Bachelors_Higher[df.binary_boulder == 0].mean()
MoE = t.ppf(0.975, degrees) * std_N1N2 * sqrt(1/N1 + 1/N2)
    # I'm getting a margin or error == 0 ... caused by last multiplied portion ... unsure how to address that

print('The results of the independent t-test are: \n\tt-value = {:4.5f}\n\tp-value = {:4.5f}'.format(ind_t_test[0],ind_t_test[1]))
print ('\nThe difference between groups is {:3.1f} [{:3.1f} to {:3.1f}] (mean [95% CI])'.format(diff_mean, diff_mean - MoE, diff_mean + MoE))

#######################################   END STUDENT CODE   #####################################################
```
